```python
from GridWorldPI import GetOutGame
import numpy as np

xdim = 20
ydim = 20
terminals = np.random.randint(low=0, high=xdim*ydim-1, size=10)
game = GetOutGame(xdim=xdim, ydim=ydim, terminals=terminals)

for gamma in [1,0.95,0.9,0.8,0.7,0.6,0.5,0,-0.01,1.25,2,3,10]:
    sv = game.get_state_values(gamma=gamma)
    game.print_state_values(sv)
```
